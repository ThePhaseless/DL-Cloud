
### 6. Utwórz dwa losowe tenosry o kształcie `(2, 3)` i wyślij  do GPU (będzie potrzebny  dostęp do GPU). Ustaw `torch.manual_seed(1234)` podczas tworzenia tensorów (nie musi to być losowe ziarno GPU). Wynik powinien wyglądać mniej więcej tak:

```
Device: cuda
(tensor([[0.0290, 0.4019, 0.2598],
         [0.3666, 0.0583, 0.7006]], device='cuda:0'),
 tensor([[0.0518, 0.4681, 0.6738],
         [0.3315, 0.7837, 0.5631]], device='cuda:0'))
```

````python
# Ustaw losowe ziarno
torch.manual_seed(1234)

# Sprawdź dostęp do GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Utwórz dwa losowe tensory na GPU
tensor1 = torch.rand(2, 3, device=device)
tensor2 = torch.rand(2, 3, device=device)

tensor1, tensor2
````

Wynik powinien być:

```
tensor([[[[0.5349, 0.1988, 0.6592, 0.6569, 0.2328, 0.4251, 0.2071, 0.6297,
           0.3653, 0.8513]]]]) torch.Size([1, 1, 1, 10])
tensor([0.5349, 0.1988, 0.6592, 0.6569, 0.2328, 0.4251, 0.2071, 0.6297, 0.3653,
```

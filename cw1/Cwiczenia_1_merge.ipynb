# Reshape

```python
A = np.arange(12)               # tworzy tablicę NumPy z liczbami od 0 do 11
A, A.shape                         # pokazuje kształt tablicy (12 elementów w 1 wymiarze)
```

```python
s = pd.Series(A.tolist())       # konwersja tablicy NumPy na listę i utworzenie obiektu Series
s.iloc[1]                       # pobranie elementu o indeksie 1 (drugi element serii)
```

```python
s                               # wyświetlenie całej serii pandas
```

Jedną z najczęściej używanych funkcji w Numpy jest reshape(), która nadaje nowy kształt tablicy bez zmiany jej danych. Kiedy używasz reshape(-1, 1), prosisz numpy o zmianę kształtu tablicy z 1 kolumną i tyloma wierszami, ile potrzeba, aby pomieścić dane. Wynikiem tej operacji będzie tablica 2D o kształcie (n, 1), gdzie n to liczba elementów w oryginalnej tablicy.

```python
C = A.reshape(-1, 1)            # zmiana kształtu tablicy na kolumnę (12 wierszy, 1 kolumna)
C                               # wyświetlenie tablicy C
C.shape                         # sprawdzenie nowego kształtu tablicy C
```

reshape(1, -1) przekształca tablicę z 1 wierszem i tyloma kolumnami, ile potrzeba, aby pomieścić dane. Wynikiem tej operacji będzie tablica 2D o kształcie (1, n), gdzie n to liczba elementów w oryginalnej tablicy.

```python
B = A.reshape(1, -1)            # zmiana kształtu tablicy na wiersz (1 wiersz, 12 kolumn)
B                               # wyświetlenie tablicy B
```

Kluczową różnicą między reshape(-1, 1) i reshape(1, -1) jest kształt wynikowej tablicy. reshape(-1, 1) skutkuje tablicą z pojedynczą kolumną i wieloma wierszami (wektor kolumn), podczas gdy reshape(1, -1) skutkuje tablicą z pojedynczym wierszem i wieloma kolumnami (wektor wierszy).

```python
D = A.reshape(3, 4)             # zmiana kształtu tablicy na macierz 3x4
D                               # wyświetlenie macierzy D
```
